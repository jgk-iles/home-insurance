## Occupation and Education Dictionaries

Just using this notebook to create dictionaries for score conversions for the education and occupation fields.

```python
import pandas as pd
import pickle
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import joblib
```

```python
mortgage_df = pd.read_csv("../data/Mortgage.csv")
campaign_df = pd.read_csv("../data/Campaign.csv")
```

```python
campaign_df.head()
```

```python
job_levels = campaign_df[['job_title', 'occupation_level']]
```

```python
job_lookup = job_levels.groupby('job_title').mean()
```

```python
job_lookup.loc['Data Scientist']
```

```python
job_dict = job_lookup.to_dict()['occupation_level']
```

```python
with open("../outputs/job_dict.pkl", "wb") as handle:
    pickle.dump(job_dict, handle)
```

```python
education_levels = campaign_df[['education', 'education_num']]
education_levels.head()
```

```python
education_lookup = education_levels.groupby('education').mean()
```

```python
education_lookup
```

```python
education_dict = education_lookup.to_dict()['education_num']
```

```python
with open("../src/edu_dict.pkl", "wb") as handle:
    pickle.dump(education_dict, handle)
```
